```
from matplotlib import pyplot as pyplot
x=[1,2,4,5]
y=[2,6,3,6]

plt.plot(x,y)
plt.title('Graph')
plt.xlabel('X axis')
plt.ylabel('Y axis')

plt.show()
```

```
from matplotlib import pyplot as pyplot
from matplotlib import style

style.use('ggplot')

x1=[1,2,4,5]
y1=[2,6,3,6]

x2=[2,3,4,5]
y2=[10,8,14,12]

plt.plot(x1,y1,'g',label='line 1',linewidth=5)
plt.plot(x2,y2,'c',label='line 2',linewidth=2)

plt.title('Style Graph')
plt.xlabel('X axis')
plt.ylabel('Y axis')

plt.legend()

plt.grid(True,color='k')
plt.show()
```
